## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
is_invertible="""We can verify if matrix T is a square matrix and has full rank to see if it is invertible. It is invertible if it is a square matrix with full rank. If a square matrix's
determinant is not zero, it has complete rank.Let's determine T's inverse Tinv after first determining whether it is invertible. The original image will then be recovered using Tinv by
performing the opposite of the scrambling procedure."""

print(is_invertible)
print()

# Check if T is invertible
if np.linalg.matrix_rank(T) == T.shape[0]:
    print("Matrix T is invertible.")
    # Compute the inverse of T
    Tinv = np.linalg.inv(T)

    # Perform the inverse scrambling operation to retrieve the original image
    original_image = np.matmul(Tinv,transformed_image)

    plt.imshow(original_image, cmap='gray')  # Assuming it's grayscale
    plt.title('Retrieved Original Image')
    plt.axis('off')
    plt.show()
else:
    print("Matrix T is not invertible.")
```

## Part 8: Extra Points

* We want to perform the same operations but on the columns of the original image. How would you do the matrix algebra? Plot the scrambled image for the proof of concept.


```python
# Get the number of columns in the original image
num_cols = image.shape[1]

# Create an identity matrix with the same number of columns
identity_matrix_cols = np.eye(num_cols)

# Shuffle the column indices
shuffled_cols = np.arange(num_cols)
np.random.shuffle(shuffled_cols)

# Create the T_columns matrix by rearranging the columns of the identity matrix
T_columns = identity_matrix_cols[:, shuffled_cols]
immg=np.matmul(Tinv,transformed_image)

# Scramble the original image by multiplying it by T_columns
scrambled_image = np.matmul(immg, T_columns)

# Plot the scrambled image
plt.imshow(scrambled_image)
plt.title('Scrambled Image (Columns)')
plt.axis('off')
plt.show()

# To retrieve the original image, compute the inverse of T_columns
T_columns_inv = np.linalg.inv(T_columns)

# Multiply the scrambled image by the inverse of T_columns to retrieve the original image
retrieved_image = np.matmul(scrambled_image,T_columns_inv)

# Plot the retrieved original image
plt.imshow(retrieved_image)
plt.title('Retrieved Original Image (Columns)')
plt.axis('off')
plt.show()
```

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image.

```python
# Get the number of rows and columns in the original image
num_rows, num_cols, num_channels = image.shape

# Create an identity matrix with the same number of rows and columns
identity_matrix_rows = np.eye(num_rows)
identity_matrix_cols = np.eye(num_cols)

# Shuffle the row and column indices
shuffled_rows = np.arange(num_rows)
shuffled_cols = np.arange(num_cols)
np.random.shuffle(shuffled_rows)
np.random.shuffle(shuffled_cols)

# Create the T_rows and T_columns matrices by rearranging the identity matrices
T_rows = identity_matrix_rows[shuffled_rows, :]
T_columns = identity_matrix_cols[:, shuffled_cols]

# Scramble the original image by multiplying it by T_rows and T_columns
scrambled_image = np.matmul(np.matmul(T_rows, immg), T_columns)

# Plot the scrambled image
plt.imshow(scrambled_image)
plt.title('Scrambled Image')
plt.axis('off')
plt.show()

# To retrieve the original image, compute the inverse of T_rows and T_columns
T_rows_inv = np.linalg.inv(T_rows)
T_columns_inv = np.linalg.inv(T_columns)

# Multiply the scrambled image by the inverses of T_columns and T_rows to retrieve the original image
retrieved_image = np.matmul(np.matmul(T_rows_inv, scrambled_image), T_columns_inv)

# Plot the retrieved original image
plt.imshow(retrieved_image)
plt.title('Retrieved Original Image')
plt.axis('off')
plt.show()
```
